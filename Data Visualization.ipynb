Visualization 1
Top 20 products by Total Price
Top 20 products by ordered quantity

```python
sns.set_style('whitegrid')
HighestOrderedProductsByQuantity = data.groupby('StockCode')['Quantity'].agg('sum').sort_values(ascending=False)[0:20]
HighestOrderedProductsByTotalPrice = data.groupby('StockCode')['FinalPrice'].agg('sum').sort_values(ascending=False)[0:20]
fig,axs = plt.subplots(nrows=2, ncols=1, figsize = (12,12))
plt.subplots_adjust(hspace = 0.3)
fig.suptitle('Best Selling Products by Quantity and Revenue', fontsize=15, x = 0.4, y = 0.98)
sns.barplot(x=HighestOrderedProductsByQuantity.index, y=HighestOrderedProductsByQuantity.values, ax= axs[0]).set(xlabel='Stock Code')
axs[0].set_title('Quantity Sold', size=12, fontweight = 'bold')
sns.barplot(x=HighestOrderedProductsByTotalPrice.index, y=HighestOrderedProductsByTotalPrice.values, ax= axs[1]).set(xlabel='Stock Code')
axs[1].set_title('Aggregate Amount received', size=12, fontweight = 'bold')
plt.show()
```

```python
palette_color = sns.color_palette('pastel')
plt.pie(HighestOrderedProductsByQuantity.values, labels=HighestOrderedProductsByQuantity.index, colors=palette_color)
plt.show()
```

```python
palette_color = sns.color_palette('pastel')
plt.pie(HighestOrderedProductsByTotalPrice.values, labels=HighestOrderedProductsByTotalPrice.index, colors=palette_color)
plt.show()
```

```python
#plotting the qunatity vs unitprice
QuantityAgainstPrice = sns.jointplot(x="UnitPrice", y="Quantity", data = data[data.FinalPrice>0], height = 7)
QuantityAgainstPrice.fig.suptitle("Unit Price and Quantity Comparison", fontsize = 15, y = 1.2)
plt.show()
```

Plot weekly and yearly sales data for non-returned items

```python
SalePerWeek = data[data['Quantity']>0].Quantity.resample('W').sum()

```

```python
#resampling to get the weekly sales and returns
SalesPerWeek = data[data['Quantity']>0].Quantity.resample('W').sum()
ReturnsPerWeek = data[data['Quantity']<0].Quantity.resample('W').sum().abs()
#creating the subplot
fig,[ax1, ax2] = plt.subplots(nrows=1,ncols=2, figsize = (15,5))
SalesPerWeek.plot(ax=ax1).set(xlabel="Month", ylabel="Quantity")
ax1.set_title("Weekly Sales Quantity", fontsize = 15)
ReturnsPerWeek.plot(ax=ax2).set(xlabel="Month", ylabel="Quantity")
ax2.set_title("Weekly Returns Quantity", fontsize = 15)
plt.show()
```

```python
#Visualizing transactions across countries
SalesPerCountry = data[ data.Quantity > 0].groupby('Country')['Quantity'].sum()
ReturnsPerCountry = data[ data.Quantity < 0].groupby('Country')['Quantity'].sum().abs()
#creating the subplot
fig, [ax1,ax2] = plt.subplots(nrows=2,ncols=1,figsize=(10,14))
SalesPerCountry.plot(kind='bar', ax=ax1).set(ylabel = 'Quantity',xlabel='')
ax1.set_title('Sales', size=12, fontweight = 'bold')
ReturnsPerCountry.plot(kind='bar', ax=ax2).set(ylabel = 'Quantity',xlabel='')
ax2.set_title('Returns', size=12, fontweight = 'bold')
plt.suptitle('Sales in Foreign Countries', fontsize = 15)
plt.subplots_adjust(hspace = 0.6)
plt.show()
```

Since UK had surpassed other countries in sales and returns by a large margin, we decided to visualize without taking into consideration UK data

```python
sns.set_style('whitegrid')
SalesPerCountry = data[ (data.Quantity > 0) & (data.Country != 'UNITED KINGDOM')].groupby('Country')['Quantity'].sum()
ReturnsPerCountry = data[ ( data.Quantity < 0 ) & (data.Country != 'UNITED KINGDOM')].groupby('Country')['Quantity'].sum().abs()
fig, [ax1,ax2] = plt.subplots(nrows=2,ncols=1,figsize=(10,14))
SalesPerCountry.plot(kind='bar', ax=ax1).set(ylabel = 'Quantity',xlabel='')
ax1.set_title('Sales', size=12, fontweight = 'bold')
ReturnsPerCountry.plot(kind='bar', ax=ax2).set(ylabel = 'Quantity',xlabel='')
ax2.set_title('Returns', size=12, fontweight = 'bold')
plt.suptitle('Sales in Foreign Countries', fontsize = 15)
plt.subplots_adjust(hspace = 0.6)
plt.show()
```

A heatmap to show the revenue variation by days of week for each month

```python

revenue_data = data.pivot_table(index = 'InvoiceMonth',columns = 'Day of week', values = 'FinalPrice', aggfunc='sum')
plt.figure(figsize = (10,6))
sns.heatmap(revenue_data, cmap="YlGnBu").set(xlabel='', ylabel='')
plt.title('Revenue per Month and Day of Week', fontsize = 15)
plt.show()
```
