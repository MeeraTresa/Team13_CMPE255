By identifying the association rules between them, we hope to identify the goods that were frequently purchased in this section.


We need to convert the data into standard format with one-hot encoding in which we will group the data by transactions and the names of the products and then unstack it so in each row we would have a transaction and each column would show if a specific product exists in that transaction. This will allow us to use apriori and association rules (from the mlxtend library).
The following table shows the products that were most frequently purchased together after the code has been run:

```python
#importing necessary libraries
import pandas as pd 
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules

#importing the dataset
clean_dataset = 'data_cleaned.csv'
cleaned_data = pd.read_csv(clean_dataset, index_col = 'InvoiceDate')
cleaned_data.index = pd.to_datetime(cleaned_data.index, format = '%Y-%m-%d %H:%M')

#converting the data into the standard form
basket = cleaned_data.loc[(cleaned_data['Quantity']>0) ,['InvoiceNo','Description','Quantity']]
basket = basket.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack(fill_value=0)
basket = (basket > 0)

#finding frequent itemsets and association rules
frequent_itemsets = apriori(basket, min_support=0.0325, use_colnames=True)
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold=0.5)
rules
```
